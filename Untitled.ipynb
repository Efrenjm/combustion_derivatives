La ecuaci√≥n qu√≠mica queda determinada de la siguiente manera:

>$$C_{\alpha}H_{\beta} +  \lambda(\alpha + \beta/4) O_{2} \longrightarrow n_{1}CO_{2} +  n_{2}H_{2}O(g) +  n_{3}CO +  n_{4}O_{2} + n_{5} H_{2}$$\
>$$CH_{4} +  2\lambda O_{2} \longrightarrow n_{1}CO_{2} +  n_{2}H_{2}O(g) +  n_{3}CO +  n_{4}O_{2} + n_{5} H_{2}$$

Para lograr el equilibrio de la ecuaci√≥n se requiere determinar las ecuaciones que gobiernan el sistema. Estas ecuaciones signfican una igualdad entre los moles de reactivos y los moles de producto para cada uno de los elementos de la reacci√≥n. De esta manera podemos obtener 3 ecuaciones para $C, H$ y $O$.
- Para C: 
    >$$ 1 = n_{1} + n_{3} $$
- Para H:
    >$$ 4 = 2n_{2} + 2n_{5} $$
- Para O:
    >$$ 4\lambda = 2n_{1} + n_{2} + n_{3} + 2n_{4} $$

Para este momento contamos con 3 ecuaciones y 5 inc√≥gnitas. Para poder resolver este sistema de ecuaciones es necesario, entonces, conocer un poco m√°s sobre el sistema en el cual se est√° llevando a cabo la combusti√≥n, es decir, si existe alguna disociaci√≥n de sus productos en equilibrio. Esto depender√° de cada apartado siguiente.


Una vez que la reacci√≥n haya sido balanceada se proceder√° a buscar la temperatura de flama adiab√°tica, la cual es la temperatura para la que no existen cambios entre el calor de los reactivos y el calor de los productos y por lo tanto, cumple que:

> $$H_R = H_P$$

Que a su vez se puede definir como:

> $$\sum_{R}n_{i}(\bar{h^{o}_f} + \bar{\Delta h})_i = \sum_{P}n_{e}(\bar{h^{o}_f} + \bar{\Delta h})_e$$

La entalp√≠a de los reactivos toma el valor de la entalp√≠a de formaci√≥n del combustible, ya que por convenci√≥n, la entalp√≠a de formaci√≥n de los elementos en su forma est√°ndar tienen un valor de cero, es decir, tanto la entalp√≠a de formaci√≥n del O<sub>2</sub> puro como la del aire es igual a 0 y por lo tanto:

>$$H_R = (\bar{h^{o}_f})_{CH_4} = -74,850kJ/kmol$$

Entonces, para cada uno de los siguientes casos se cumple que la temperatura de flama adiab√°tica se obtiene cuando las respectivas sumas de las entalp√≠as de los productos sean igual a $-74,850kJ/kmol$.

### Resoluci√≥n del sistema de ecuaciones
Se crea el sistema de ecuaciones, el cual se resolver√° por el m√©todo de Gauss-Jordan. La funci√≥n de balanceo es espec√≠fica para cada tipo de reacci√≥n.

Para poder resolver el sistema de ecuaciones es necesario utilizar una temperatura de prueba arbitraria, en este caso se ha seleccionado $T_P = 2300$

Una vez balanceada la ecuaci√≥n qu√≠mica, se analiza si se cumple que para la temperatura $T_P$, $H_R = H_P = ‚àí74,850ùëòùêΩ/ùëòùëöùëúùëô$. La entalp√≠a de los productos se puede encontrar con la siguiente f√≥rmula.

>$$H_P = \sum_{P}n_{e}(\bar{h^{o}_f} + \bar{\Delta h})_e$$\
$$= n_{1}(\bar{h^{o}_f}+\bar{\Delta h})_{CO_2} + n_{2}(\bar{h^{o}_f}+\bar{\Delta h})_{CO} + n_{3}(\bar{h^{o}_f}+\bar{\Delta h})_{H_{2}O} + n_{4}(\bar{h^{o}_f}+\bar{\Delta h})_{O_2} + n_{5}(\bar{h^{o}_f}+\bar{\Delta h})_{H_2}$$

Se procede a analizar individualmente

>$$(\bar{h^{o}_f}+\bar{\Delta h})_{CO_2} = n_{1}(-393,520kJ/kmol + \bar{\Delta h}_{CO_{2}})$$\
$$(\bar{h^{o}_f}+\bar{\Delta h})_{CO} = n_{2}(-110,530kJ/kmol + \bar{\Delta h}_{CO})$$\
$$(\bar{h^{o}_f}+\bar{\Delta h})_{H_{2}O} = n_{3}(-241,820kJ/kmol + \bar{\Delta h}_{H_{2}O})$$\
$$(\bar{h^{o}_f}+\bar{\Delta h})_{O_2} = n_{4}(0kJ/kmol + \bar{\Delta h}_{H_{2}O})$$\
$$(\bar{h^{o}_f}+\bar{\Delta h})_{H_2} = n_{5}(0kJ/kmol +\bar{\Delta h}_{H_{2}})$$


A partir de esto, procedemos a resolver la ecuaci√≥n de forma iterativa para encontrar el caso para el que estas condiciones se cumplan en equilibrio.

### Librer√≠as importantes y obtenci√≥n de datos globales.
Las siguientes librer√≠as ser√°n fundamentales para el desarrollo de todo el proyecto. Se importan las librer√≠as de manejo de gr√°ficos, manejo de funciones matem√°ticas avanzadas y manejo de acceso a archivos. Adem√°s se definen una funci√≥n de resoluci√≥n de sistemas de ecuaciones, una funci√≥n de interpolaci√≥n, una funci√≥n de m√∫ltiples interpolaciones, as√≠ como la funci√≥n de la obtenci√≥n de la temperatura de flama adiab√°tica y de suma de sus valores de entalp√≠as parciales.

Para lambda se han seleccionado 80 valores linealmente distribuidos entre 0.5 y 2.5. El criterio de aceptaci√≥n para la resoluci√≥n de este sistema de ecuaciones es $|H_P-H_R| < \epsilon$ donde $\epsilon = 10^{-8}$.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

datosh = pd.read_csv('Entalpias.csv')   # Tabla de valores de entalp√≠as 
datose = pd.read_csv('Entropias.csv')   # Tabla de valores de entrop√≠as 

#Se asignan variables para el f√°cil manejo de los datos.
temp = datosh['Temperatura'].values
hAir = datosh['hAire'].values             ;eAir = datose['Air'].values
hN2  = datosh['hN2'  ].values             ; eN2 = datose['N2' ].values
hO2  = datosh['hO2'  ].values             ; eO2 = datose['O2' ].values
hH2  = datosh['hH2'  ].values             ; eH2 = datose['H2' ].values
hCO  = datosh['hCO'  ].values             ; eCO = datose['CO' ].values
hCO2 = datosh['hCO2' ].values             ;eCO2 = datose['CO2'].values
hH2O = datosh['hH2O' ].values             ;eH2O = datose['H2O'].values
hNO  = datosh['hNO'  ].values             ; eNO = datose['NO' ].values

#Entalp√≠as de formaci√≥n para mol√©culas de inter√©s.
hfCH4 = -74850;     hfO2 = 0;           hfH2 = 0;           hfN2 = 0;       hfAir=0  
hfCO = -110525;     hfCO2 = -393512;    hfH2O = -241824;    hfNO = 90800    

#Valor de la entalp√≠a de los reactivos
Hr = -74520     

# Criterio de aceptaci√≥n para Hp=Hr
cauchy = 1e-4

# Valores para lambda, asegur√°ndonos de que exista lambda = 1
lamb = np.linspace(0.5,2,150)
if np.where(lamb==1)[0].size == 0:
    uno, = np.where(lamb<1)
    lamb = np.insert(lamb,uno[-1]+1,1)

#Soluci√≥n de sistema de ec. por Gauss-Jordan
def gauss(matriz):        
    n,m = matriz.shape
    for i in range(n):
        matriz[i,:] = matriz[i,:]/matriz[i,i]
        for j in range(n):
            if j==i:
                continue
            matriz[j,:] -= matriz[i,:]*matriz[j,i]
    return matriz

def interpolacion(y1,y0,x1,x0,x):     #Interpolaci√≥n lineal
    return ((y1-y0)/(x1-x0))*(x-x0)+y0

#Interpolaci√≥n de varios elementos a la vez
def interpolaciones(lowindex,refmatrix,val,*matrices)  :
    interpol = np.arange(0)
    xinf = refmatrix[lowindex]; xsup = refmatrix[lowindex+1]
    for elem in matrices :
        yinf = elem[lowindex]; ysup = elem[lowindex +1]
        yi = ((ysup-yinf)/(xsup-xinf))*(val-xinf)+yinf
        interpol = np.append(interpol,yi)
    return interpol


# Obt√©n la suma de las entalp√≠as parciales de los compuestos
def sumDeltah(temper, coef, *productos, temp=temp):
    Dh_pi = np.arange(0)
    Tindex, = np.where(temp == temper)  #Busca la temperatura en las tablas
        
    if Tindex.size > 0 :       #Si se encontr√≥, anexa los valores de tablas para cada compuesto
        for compuesto in productos:
            Dh_pi = np.append(Dh_pi, compuesto[int(Tindex)])
                
    else:   #Si no se encontr√≥ en las tablas, 
            
            #Encuentra √≠ndice del valor de temperatura inmediato inferior a la temperatura buscada en la tabla
        infval = temp[np.bitwise_and(temp>temper-101,temp<temper)] 
        (Tindex,), = np.where(temp == infval[0])
            
            #interpola los valores de h para cada compuesto y an√©xalo a la matriz de entalp√≠as
        for compuesto in productos:
            hi = interpolaciones(Tindex,temp,temper,compuesto)
            Dh_pi = np.append(Dh_pi, hi)
                
    # Retorna la suma de los productos de la entalp√≠a con sus respectivos coeficientes
    return np.dot(coef,Dh_pi) 

# Funci√≥n iterativa para encontrar el valor de la temperatura de flama adiab√°tica.
def flame(balance,Hr,T,lamb,hFormi,*productos,temp=temp,cauchy=cauchy):    #Funci√≥n para la obtenci√≥n de la temperatura de flama adiab√°tica
    cont = 0; maximo = 50    # Contador de seguridad para el loop.
    
    (Ti,), = np.where(temp==T)  # Obten el √≠ndice de la temperatura propuesta en las tablas

    coef = balance(lamb,T)            # Balancea la ecuaci√≥n qu√≠mica
    tothForm = np.dot(hFormi,coef)    # Suma las entalp√≠as de formaci√≥n
    Hp = sumDeltah(T,coef,*productos) + tothForm    # Calcula la entalp√≠a de los productos
    
        #El loop se repite mientras Hp y Hr sean diferentes y mientras el contador no llegue al m√°ximo.
    while np.bitwise_and(abs(Hp - Hr) > cauchy, cont < maximo):  
            # Se guardan los valores obtenidos en el ciclo anterior (o los iniciales)
        lastHp = Hp        ;lastTi = Ti           ;lastT = T 
        
        Ti = Ti+1 if Hp<Hr else Ti-1    # Usa la pr√≥xima o la anterior temperatura de la tabla
        if np.bitwise_or(Ti < 0,Ti>=temp.size ):  # Si se ha llegado a los l√≠mites de la tabla, sal del loop
            break
        T  = temp[Ti]   
        
            #Calcula Hp para el nuevo valor de T
        coef = balance(lamb,T)
        tothForm = np.dot(hFormi,coef)
        Hp = sumDeltah(T,coef,*productos) + tothForm   
                
            # Si hubo un cambio de Hp-Hr con respecto a 0 entre este ciclo y el anterior interpola la temperatura
        if np.bitwise_and(Hp < Hr, lastHp > Hr) | np.bitwise_and(Hp > Hr, lastHp < Hr):
                #Interpola el valor de T entre el valor anterior y el reci√©n obtenido
            T = interpolacion(T,lastT,Hp,lastHp,Hr) #Interpola la temperatura 
            
                #Calcula nuevamente Hp para el valor de interpolado de T
            coef = balance(lamb,T)
            tothForm = np.dot(hFormi,coef)
            Hp = sumDeltah(T,coef,*productos) + tothForm

        cont += 1
    return T,Hp,coef
```

## 1.1 Combusti√≥n completa (sin disociaci√≥n).

En este primer caso, se habla de la combusti√≥n sin disociaci√≥n de sus productos, y por lo tanto, eso limita un poco la estructuraci√≥n del sistema, sin embargo, podemos asumir un par de situaciones de estado a partir de esto.

Para mezclas ricas con poca oxigenaci√≥n, es decir $\lambda < 1$, el $O_2$ se mezclar√≠a completamente con el metano para producir la reacci√≥n y por lo tanto, no encontrar√≠amos moles de esta sustancia entre los productos. Mientras que para mezclas d√©biles en donde $\lambda > 1$ se ha acordado que no existe disociaci√≥n, entonces, sustancias que son resultado de tales reacciones de disociaci√≥n como el $H_2$ o el $CO$ no aparecer√≠an entre los productos. De igual forma, en una combusti√≥n estequim√©trica, es decir, $\lambda = 1$ se cumplir√≠an ambas condiciones se√±aladas anteriormente.

- Para mezclas ricas.
>$$ \lambda < 1:   \hspace{7cm}  \{n_{O_2}= n_4\} = 0 $$ 
- Para mezclas d√©biles.
>$$ \lambda > 1:  \hspace{4.5cm}    \{n_{CO}=n_3\}= \{n_{H_2} = n_5\} = 0 $$
- Para mezclas estequiom√©tricas 
>$$ \lambda = 1:   \hspace{2cm}   \{n_{CO}=n_3\}=\{n_{O_2}= n_4\}= \{n_{H_2} = n_5\} = 0 $$

- **Mezcla rica $(\lambda<1)$**

A pesar de que se han asumido ciertas circunstancias, las mezclas ricas a√∫n necesitar√≠an de una ecuac√≠n m√°s para poder converger en una soluci√≥n. Esta √∫ltima ecuaci√≥n puede ser obtenida a partir de la ecuaci√≥n de equilibrio de agua ‚Äì gas.

>$$ CO_2 + H_2 \rightleftharpoons CO + H_2O$$

Esta es una ecuaci√≥n conocida que ocurre naturalmente a altas temperaturas y que al ser una ecuaci√≥n de disociaci√≥n en equilibrio cumple que las irreversibilidades del sistema son iguales a cero.

>$$ \overline{dG_{T,P}} = \upsilon_{CO}\overline{G_{CO}} + \upsilon_{H_2O}\overline{G_{H_2O}} - \upsilon_{CO_2}\overline{G_{CO_2}} - \upsilon_{H_2}\overline{G_{H_2}} = 0$$

Y por lo tanto, esta proporci√≥n de equilibrio debe de cumplirse para los valores estequiom√©tricos del sistema, lo que nos otorga una nueva ecuaci√≥n.

>$$n_3\overline{G_{CO}} + n_2\overline{G_{H_2O}} - n_1\overline{G_{CO_2}} - n_5\overline{G_{H_2}} = 0$$

De esta manera, ya se pueden formar sistemas de ecuaciones que converjan.

\end{equation}\

>\begin{equation}
    \tag{Ecuaci√≥n 1.4}

\end{equation}\
>\begin{equation}
    \tag{Ecuaci√≥n 1.5}
    n_3\overline{G_{CO}} + n_2\overline{G_{H_2O}} - n_1\overline{G_{CO_2}} - n_5\overline{G_{H_2}} = 0
\end{equation}

### Funci√≥n de balanceo para el apartado 1.1.
Las ecuaciones de balanceo se enlistan como una matriz la cual se encuentra en el orden:
>$$[n1, n2, n3, n4, n5, \text{valor de soluci√≥n} ]$$

Donde la ecuaci√≥n qu√≠mica que rige este sistema es.
>$$CH_{4} +  2\lambda O_{2} \longrightarrow n_{1}CO_{2} +  n_{2}H_{2}O(g) +  n_{3}CO +  n_{4}O_{2} + n_{5} H_{2}$$

```python
# Funci√≥n de resoluci√≥n del sistema de ecuaciones para el ejercicio 1.1
def balance11(lamb,temper,temp=temp,eH2=eH2,eCO2=eCO2,eH2O=eH2O,eCO=eCO):

    Tindex, = np.where(temp == temper)
    if Tindex.size>0 :              #Si el valor de temperatura es igual a alguno de la tabla toma valores
        Tindex = int(Tindex)    
        GH2 = eH2[Tindex]; GCO2 = eCO2[Tindex]; GH2O = eH2O[Tindex]; GCO = eCO[Tindex]
        
    else:                           #Si no est√° en la tabla, interpola entre valores
        mask = temp[np.bitwise_and(temp>temper-101,temp<temper)]
        (Tindex,), = np.where(temp == mask[0])
        GH2, GCO2, GH2O, GCO = interpolaciones(Tindex,temp,temper,eH2,eCO2,eH2O,eCO)
    
    #Modelado del sistema de ecuaciones: n1 = nCO2, n2 = nH2O, n3 = nCO, n4 = nO2, n5 = nH2
                        #[   n1,   n2,  n3, n4,   n5,sol]

    if  lamb<1:     #Sin producci√≥n de O2 y con equilibrio agua gas  
        coef = np.array([[    1,    0,   1,  0,    0, 1],        # Balance de C
                         [    0,    1,   0,  0,    1, 2],        # Balance de H
                         [    2,    1,   1,  0,    0, 4*lamb],   # Balance de O2
                         [    0,    0,   0,  1,    0, 0],        # Ausencia de O2 en los productos
                         [-GCO2, GH2O, GCO,  0, -GH2, 0]])       # Ecuaci√≥n de equilibrio agua-gas
    
    elif lamb ==1:  #Sin producci√≥n de O2, CO ni H2
        coef = np.array([[    1,    0,   0,  0,    0, 1],           # Balance de C
                         [    0,    1,   0,  0,    0, 2],           # Balance de H
                         [    0,    0,   1,  0,    0, 0],           # Ausencia de O2 en los productos
                         [    0,    0,   0,  1,    0, 0],           # Ausencia de CO en los productos
                         [    0,    0,   0,  0,    1, 0]])          # Ausencia de H2 en los productos

    elif lamb>1:   #Sin producci√≥n de CO ni H2
        coef = np.array([[    1,    0,   0,  0,    0, 1],           # Balance de C
                         [    0,    1,   0,  0,    0, 2],           # Balance de H
                         [    0,    0,   1,  0,    0, 0],           # Ausencia de O2 en los productos
                         [    0,    0,   0,  1,    0, 2*(lamb-1)],  # Balance de O2
                         [    0,    0,   0,  0,    1, 0]])          # Ausencia de H2 en los productos

    coef = gauss(coef)
    return coef[:,5]

tablaDeSolucion11 = pd.DataFrame()
hiForm = np.array([hfCO2,hfH2O,hfCO,hfO2,hfH2])

T = 2300   #Temperatura de prueba
Tfa11 = np.arange(0)
solucionQuimica = np.arange(0)
for lambd in lamb:
    
        # Calcula la temperatura de flama adiab√°tica y la entalp√≠a de los pruductos a tal temperatura
    temperatura,Hp,coef = flame(balance11, Hr, T, lambd, hiForm , hCO2,hH2O,hCO,hO2,hH2)

        #Guarda los valores
    Tfa11 = np.append(Tfa11,round(temperatura,2)) if abs(Hp-Hr)<cauchy else np.append(Tfa11,np.nan)
    solucionQuimica = np.append(solucionQuimica,coef) if abs(Hp-Hr)<cauchy else np.append(solucionQuimica,np.nan)

solucionQuimica = solucionQuimica.reshape(-1,5)
tablaDeSolucion11['Lambda'] = lamb
tablaDeSolucion11['Temperatura_De_Flama_Adiabatica'] = Tfa11
for i in range (5) :
    tablaDeSolucion21['n'+str(i+1)] = solucionQuimica[:,i]
print('listo')
```

### 1.2 Disociaci√≥n de los productos usando la Ecuaci√≥n 1.
En esta ocasi√≥n se deber√° tomar en cuenta que durante la combusti√≥n existe una disociaci√≥n del tipo:

>\begin{equation}
    2 CO_2  \rightleftharpoons 2CO + O_2
\end{equation}

En este caso, se cumple la condici√≥n de equilibrio de la que se habl√≥ con la ecuaci√≥n de equilibrio agua-gas.
>$$ \overline{dG_{T,P}} = 2\upsilon_{CO}\overline{G_{CO}} + \upsilon_{O_2}\overline{G_{O_2}} - 2\upsilon_{CO_2}\overline{G_{CO_2}} = 0$$

Y por lo tanto, en nuestro sistema de ecuaciones se puede anexar esta nueva ecuaci√≥n:
>$$ 2n_3\overline{G_{CO}} + n_4\overline{G_{O_2}} - 2n_1\overline{G_{CO_2}} = 0$$

Entonces ahora contamos con un sistema de ecuaciones con 5 inc√≥gnitas y 4 ecuaciones, por lo que se anexan un par de condiciones para poder completar el sistema.

- Mezclas ricas ($\lambda < 1$)

Para este caso, no podemos asumir la ausencia de $O_2$ en la combusti√≥n, ya que forma parte del equilibro de la ecuaci√≥n de disociaci√≥n del $CO_2$, entonces usamos la ecuaci√≥n de equilibrio agua gas para poder completar el sistema.
>\begin{equation}
    \tag{Ecuaci√≥n 1.16}
    n_1 + n_3 = 1 
\end{equation}\
>\begin{equation}
    \tag{Ecuaci√≥n 1.17}
    n_2 + n_5 = 2 
\end{equation}\
>\begin{equation}
    \tag{Ecuaci√≥n 1.18}
    2n_1 + n_2 + n_3 + 2n_4 = 4\lambda
\end{equation}\
>\begin{equation}
    \tag{Ecuaci√≥n 1.19}
    n_3\overline{G_{CO}} + n_2\overline{G_{H_2O}} - n_1\overline{G_{CO_2}} - n_5\overline{G_{H_2}} = 0
\end{equation}\
>\begin{equation}
    \tag{Ecuaci√≥n 1.20}
    2n_3\overline{G_{CO}} + n_4\overline{G_{CO}} - 2n_1\overline{G_{CO_2}} = 0
\end{equation}

- Mezclas estequiom√©tricas y mezclas d√©biles ($\lambda ‚â• 1$)

Para el caso de las mezclas estequiom√©tricas y las mezclas d√©biles se asume que no hay producci√≥n de $H_2$, es decir, $n_5 = 0$. En este caso, tambi√©n se utilizan las tres ecuaciones de balanceo y la ecuaci√≥n de equilibrio de la disociaci√≥n de $CO_2$.


>\begin{equation}
    \tag{Ecuaci√≥n 1.21}
    n_1 + n_3 = 1 
\end{equation}\
>\begin{equation}
    \tag{Ecuaci√≥n 1.22}
    n_2 + n_5 = 2 
\end{equation}\
>\begin{equation}
    \tag{Ecuaci√≥n 1.23}
    2n_1 + n_2 + n_3 + 2n_4 = 4\lambda
\end{equation}\
>\begin{equation}
    \tag{Ecuaci√≥n 1.24}
    2n_3\overline{G_{CO}} + n_4\overline{G_{O_2}} - 2n_1\overline{G_{CO_2}} = 0
\end{equation}\
>\begin{equation}
    \tag{Ecuaci√≥n 1.25}
    n_5 = 0
\end{equation}

De esta manera se procede a ejecutar la rutina.

```python
def balance12(lamb,temper,temp=temp,eH2=eH2,eCO2=eCO2,eH2O=eH2O,eCO=eCO,eO2=eO2): 
    
    Tindex, = np.where(temp == temper) #Primero busca la temperatura en las tablas
    if Tindex.size > 0:        # Si la temperatura se encuentra tal cual en las tablas
        Tindex = int(Tindex)
        GH2O = eH2O[Tindex];    GH2 = eH2[Tindex];     GO2 = eO2[Tindex];      GCO = eCO[Tindex]
        GCO2 = eCO2[Tindex];
    else :                     # Si la temperatura no se encuentra en las tablas
            # Toma los valor inmediato inferior e interpola el valor
        mask = temp[np.bitwise_and(temp>temper-101,temp<temper)] 
        (Tindex,), = np.where(temp == mask[0])
        GO2, GCO2, GCO, GH2O, GH2 = interpolaciones(Tindex,temp,temper,eO2,eCO2,eCO,eH2O,eH2)


    #Modelado del sistema de ecuaciones: n1 = nCO2, n2 = nH2O, n3 = nCO, n4 = nO2, n5 = nH2
                        #[     n1,   n2,    n3,  n4,   n5, sol]
    if  lamb<1:
        coef = np.array([[      1,    0,     1,   0,    0,  1],       # Balance del C
                         [      2,    1,     1,   2,    0,  4*lamb],  # Balance del O 
                         [-2*GCO2,    0, 2*GCO, GO2,    0,  0],       # Ecuaci√≥n de disociaci√≥n del CO2
                         [  -GCO2, GH2O,   GCO,   0, -GH2,  0],       # Ecuaci√≥n de equilibrio agua-gas
                         [      0,    1,     0,   0,    1,  2]])      # Balance del H

    elif lamb >= 1:  # Para mezclas d√©biles y estequim√©trica no se considera producci√≥n de H2
        coef = np.array([[      1,    0,     1,   0,  0,  1],       # Balance del C
                         [      0,    1,     0,   0,  1,  2],       # Balance del H
                         [-2*GCO2,    0, 2*GCO, GO2,  0,  0],       # Ecuaci√≥n de disociaci√≥n del CO2
                         [      2,    1,     1,   2,  0,  4*lamb],  # Balance del O
                         [      0,    0,     0,   0,  1,  0]])      # Ausencia de H2

    coef = gauss(coef)
    return coef[:,5]

tablaDeSolucion12 = pd.DataFrame()
hiForm = np.array([hfCO2,hfH2O,hfCO,hfO2,hfH2])

T = 2300   #Temperatura de prueba
Tfa12 = np.arange(0)
solucionQuimica = np.arange(0)
for lambd in lamb:
    T = 2300   #Temperatura de prueba
        # Calcula la temperatura de flama adiab√°tica y la entalp√≠a de los pruductos a tal temperatura
    temperatura,Hp,coef = flame(balance12, Hr, T, lambd, hiForm , hCO2,hH2O,hCO,hO2,hH2)

        #Guarda los valores
    Tfa12 = np.append(Tfa12,round(temperatura,2)) if abs(Hp-Hr)<cauchy else np.append(Tfa12,np.nan)
    solucionQuimica = np.append(solucionQuimica,coef) if abs(Hp-Hr)<cauchy else np.append(solucionQuimica,np.empty(5))

solucionQuimica = solucionQuimica.reshape(-1,5)
tablaDeSolucion12['Lambda'] = lamb
tablaDeSolucion12['Temperatura_De_Flama_Adiabatica'] = Tfa12
for i in range (5) :
    tablaDeSolucion21['n'+str(i+1)] = solucionQuimica[:,i]
print('listo')
```

Una vez conociendo las ecuaciones que rigen este sistema, se les aplica el algoritmo.


```python
def balance21(lamb,temper,temp=temp,eH2=eH2,eCO2=eCO2,eH2O=eH2O,eCO=eCO):

    Tindex, = np.where(temp == temper)
    if Tindex.size>0 :              #Si el valor de temperatura es igual a alguno de la tabla toma valores
        Tindex = int(Tindex)
        GH2 = eH2[Tindex]; GCO2 = eCO2[Tindex]; GH2O = eH2O[Tindex]; GCO = eCO[Tindex]
        
    else:                           #Si no est√° en la tabla, interpola entre valores
        mask = temp[np.bitwise_and(temp>temper-101,temp<temper)]
        Tindex, = np.where(temp == mask[0])
        Tindex = int(Tindex)
        GH2,GCO2,GH2O,GCO = interpolaciones(Tindex,temp,temper,eH2,eCO2,eH2O,eCO)

    #Modelado del sistema de ecuaciones: n1 = nCO2, n2 = nH2O, n3 = nCO, n4 = nO2, n5 = nH2, n6 = nNO y n7 = nN2
                   #[n1,n2,n3,n4,n5,sol]
    if  lamb<1:
        coef = np.array([[    1,    0,   1,  0,    0,  0,  0,  1],                # n1 + n3 = 1
                         [    0,    1,   0,  0,    1,  0,  0,  2],                # n2 + n5 = 2
                         [    2,    1,   1,  2,    0,  1,  0,  4*lamb],           # 2n1 + n2 + n3 + 2n4 +n6 = 4*lambda
                         [    0,    0,   0,  1,    0,  0,  0,  0],                # n4 = 0 (Sin O2)
                         [-GCO2, GH2O, GCO,  0, -GH2,  0,  0,  0],                # Equilibrio agua gas
                         [    0,    0,   0,  0,    0,  1,  0,  0],       # 2n6 + n7 = 15.04*lambda
                         [    0,    0,   0,  0,    0,  0,  1,  7.52*lamb]])      # n7 = 0 (Sin N2)
                         

    
    elif lamb ==1:  #Sin producci√≥n de O2, CO, H2 ni NO
        coef = np.array([[    1,    0,   0,  0,    0,  0,  0,  1],                # 1 n1 + n3 = 1
                         [    0,    1,   0,  0,    0,  0,  0,  2],                # 2 n2 + n5 = 2
                         [    0,    0,   1,  0,    0,  0,  0,  0],                # 3 n3 = 0 (Sin CO)
                         [    0,    0,   0,  1,    0,  0,  0,  0],                # 4 n4 = 0 (Sin O2)
                         [    0,    0,   0,  0,    1,  0,  0,  0],                # 5 n5 = 0 (Sin H2)
                         [    0,    0,   0,  0,    0,  1,  0,  0],       # 6 2n6 + n7 = 15.04*lambda
                         [    0,    0,   0,  0,    0,  0,  1,  7.52]])          # 7 2n1 + n2 + n3 + 2n4 + n7 = 4

    elif lamb>1:   #Sin producci√≥n de CO, H2 ni NO
        coef = np.array([[    1,    0,   0,  0,    0,  0,  0,  1],                # 1 Balance de C
                         [    0,    1,   0,  0,    0,  0,  0,  2],                # 2 Balance de H
                         [    0,    0,   1,  0,    0,  0,  0,  0],                # 3 (Sin CO)
                         [    0,    0,   0,  1,    0,  0,  0,  2*(lamb-1)],       # 4 Balance del O
                         [    0,    0,   0,  0,    1,  0,  0,  0],                # 5 (Sin H2)
                         [    0,    0,   0,  0,    0,  1,  0,  0],                # 6 (Sin NO)
                         [    0,    0,   0,  0,    0,  0,  1,  7.52*lamb]])       # 7 Balance del N

    coef = gauss(coef)
    return coef[:,7]

tablaDeSolucion21 = pd.DataFrame()
hiForm = np.array([hfCO2,hfH2O,hfCO,hfO2,hfH2,hfNO,hfN2])

Tfa21 = np.arange(0)
solucionQuimica = np.arange(0)
for lambd in lamb:
    T = 2300
        # Calcula la temperatura de flama adiab√°tica y la entalp√≠a de los pruductos a tal temperatura
    temperatura,Hp,coef = flame(balance21, Hr, T, lambd, hiForm , hCO2,hH2O,hCO,hO2,hH2,hNO,hN2)
        # Si la temperatura calculada es aproximadamente igual a la de los reactivos, gu√°rdalo, sino, inserta un valor nulo 

    Tfa21 = np.append(Tfa21,np.round(temperatura,2)) if abs(Hp-Hr)<cauchy else np.append(Tfa21,np.nan)
    solucionQuimica = np.append(solucionQuimica,coef) if abs(Hp-Hr)<cauchy else np.append(solucionQuimica,np.empty(7))

solucionQuimica = solucionQuimica.reshape(-1,7)
tablaDeSolucion21['Lambda'] = lamb
tablaDeSolucion21['Temperatura_De_Flama_Adiabatica'] = Tfa21
for i in range (7) :
    tablaDeSolucion21['n'+str(i+1)] = solucionQuimica[:,i]
print('listo')
```
